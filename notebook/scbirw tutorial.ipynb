## importing scbirw

```python
import numpy as np
import pandas as pd
import scprep

import scbirw
```

## loading data

```python
df = pd.read_csv(
    '../data/baron_data.csv.gzip',
    header=None,
    compression='gzip'
)
df.head()
```

## Data Preprocessing

### library size normalization

```python
df = scprep.plot.plot_library_size(df)
```

### square root transform

```python
df = scprep.transform.sqrt(df)
```

## Running scBiRW

### create scbirw instance

```python
scbirw = scbirw.SCBIRW()
```

```python
?scbirw.fit_transform
```

```python
imputed = scbirw.fit_transform(
    X=df, 
    alpha=0.01, 
    Iter=10, 
)
```
